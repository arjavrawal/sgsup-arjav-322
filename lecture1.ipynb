```python
cities_data = [
    {"name": "New York", "elevation_m": 1609},
    {"name": "Denver", "elevation_m": 5280},
    {"name": "Los Angeles", "elevation_m": 285},]

elevation_threshold = 1000

print(f"Searching for first city with elevation above {elevation_threshold} meters")
for city in cities_data:
    if city["elevation_m"] > elevation_threshold:
        print(f"Found first city: {city['name']} with elevation {city['elevation_m']} meters")
        break

print(f"Processing cities with elevation above {elevation_threshold} meters")
for city in cities_data:
    if city["elevation_m"] <= elevation_threshold:
        continue
    print(f"Processing city: {city['name']} has elevation {city['elevation_m']} meters")
```
