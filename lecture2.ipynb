```python
with open("lecture2_data/coordinates.txt", "r") as file:
    lines = file.readlines()
    for line in lines:
        line = line.strip()
        if not line:
            continue
        lat_str, lon_str = line.split(",")
        lat = float(lat_str); lon = float(lon_str)
        print(f"Latitude {lat}, Longitude {lon}")
```

```python
import csv

infile = open("lecture2_data/coordinates.txt", "r")
outfile = open("lecture2_data/points.csv", "w", newline="")
csv_writer = csv.writer(outfile)
csv_writer.writerow(["Latitude", "Longitude"])
for line in infile:
    line = line.strip()
    if line:
        lat, lon = line.split(",")
        csv_writer.writerow([lat, lon])
infile.close()
outfile.close()
```

```python
import pandas as pd
from pathlib import Path

data_path = Path("lecture2_data")

population_file = data_path / "country_population.csv"

population_df = pd.read_csv(population_file)
population_largest = population_df.nlargest(5, 'Population')
population_largest
```

```python
area_file = data_path / "country_area.csv"

area_df = pd.read_csv(area_file)
area_largest = area_df.nlargest(5, 'Area')
area_largest
```

```python
merged_df = pd.merge(population_df, area_df, on='Country')
merged_df['Population_Density'] = merged_df['Population'] / merged_df['Area']
merged_df
```

```python
density_largest = merged_df.nlargest(5, 'Population_Density')
density_largest
```
